```python
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('owid-covid-data.csv')
df_india = df[df['location']=='India']
df_india_cases = df_india[['date','new_cases','total_cases']]
df_india_cases.new_cases = df_india_cases.new_cases.astype(int)
df_india_cases.total_cases = df_india_cases.total_cases.astype(int)
df_usa = df[df['location']=='United States']
df_usa_cases = df_usa[['date','new_cases','total_cases']]
df_usa_cases.new_cases = df_usa_cases.new_cases.astype(int)
df_usa_cases.total_cases = df_usa_cases.total_cases.astype(int)
new_cases_india_scaled = df_india_cases.new_cases/1000000
new_cases_usa_scaled = df_usa_cases.new_cases/1000000
total_days_india = len(df_india_cases)
day_range_india = np.arange(30, total_days_india+30)
total_days_usa = len(df_usa_cases)
day_range_usa = np.arange(22, total_days_usa+22)
plt.figure(figsize=(12, 5))
plt.plot(day_range_india, new_cases_india_scaled, color='blue')
plt.plot(day_range_usa, new_cases_usa_scaled, color='red')
plt.title('New cases for India', fontsize='24', color='black')
plt.ylabel("Number of COVID cases [millions]", fontsize=18, color='brown')
plt.xlabel("Day# (Start day - Jan 1 2020)", fontsize=18, color='green')
plt.xticks(fontsize=12, color='green')
plt.yticks(fontsize=16, color='brown')
plt.legend(('New Cases (India)', 'New Cases (USA)'), loc='upper left',shadow=True, fontsize=18)
plt.ylim([0,0.45])
plt.xlim([1,600])
plt.grid()
plt.show()
```

# **4.1.1 Plotting total cases**

```python
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np
df = pd.read_csv('owid-covid-data.csv')
df_india = df[df['location']=='India']
df_india_cases = df_india[['date','new_cases','total_cases']]
df_india_cases.new_cases = df_india_cases.new_cases.astype(int)
df_india_cases.total_cases = df_india_cases.total_cases.astype(int)
df_usa = df[df['location']=='United States']
df_usa_cases = df_usa[['date','new_cases','total_cases']]
df_usa_cases.new_cases = df_usa_cases.new_cases.astype(int)
df_usa_cases.total_cases = df_usa_cases.total_cases.astype(int)
new_cases_india_scaled = df_india_cases.new_cases/1000000
new_cases_usa_scaled = df_usa_cases.new_cases/1000000
total_cases_india_scaled = df_india_cases.total_cases/1000000
total_cases_usa_scaled = df_usa_cases.total_cases/1000000
total_days_india = len(df_india_cases)
day_range_india = np.arange(30, total_days_india+30)
total_days_usa = len(df_usa_cases)
day_range_usa = np.arange(22, total_days_usa+22)
fig, ax_1 = plt.subplots()
fig.set_size_inches(15,5)
ax_1.plot(day_range_india, new_cases_india_scaled, color='blue')
ax_1.plot(day_range_usa, new_cases_usa_scaled, color='brown')
ax_1.set_title('New and Total cases for India and United States',fontsize='24', color='black')
ax_1.set_ylabel("New COVID cases [millions]", fontsize=18, color='red')
ax_1.set_xlabel("Day# (Start day - Jan 1 2020)", fontsize=18, color='green')
ax_1.tick_params(axis='x', labelsize=12, labelcolor='green')
ax_1.tick_params(axis='y', labelsize=16, labelcolor='red')
ax_1.legend(('New Cases (India)', 'New Cases (USA)'), loc='upper left',shadow=True, fontsize=18)
ax_1.set_ylim([0,0.5])
ax_2 = ax_1.twinx()
ax_2.plot(day_range_india, total_cases_india_scaled, color='blue')
ax_2.plot(day_range_usa, total_cases_usa_scaled, color='brown')
ax_2.set_ylabel("Total COVID cases [millions]", fontsize=18, color='purple')
ax_2.tick_params(axis='x', labelsize=12, labelcolor='green')
ax_2.tick_params(axis='y', labelsize=16, labelcolor='purple')
ax_2.legend(('Total Cases (India)', 'Total Cases (USA)'), loc='center left',shadow=True, fontsize=18)
ax_2.set_ylim([0,35])
plt.grid()
plt.show()
```
